## 2) QUALITY CONTROL

```python
adata.var_names_make_unique()
adata.obs_names_make_unique()
```

```python
adata.var["mt"] = adata.var_names.str.startswith("MT-")
sc.pp.calculate_qc_metrics(adata, qc_vars=["mt"], percent_top=None, inplace=True)
adata.obs.head()
```

```python
adata.var['RIBO'] = adata.var_names.str.startswith("RPS", "RPL")
adata.var['HB'] = adata.var_names.str.startswith("^HB[^(P)]")
```

## 3. QUALITY CONTROL FILTERING

```python
adata = adata[adata.obs.n_genes_by_counts < 4000, :]
adata = adata[adata.obs.n_genes_by_counts > 300, :]
adata = adata[adata.obs.pct_counts_mt < 10, :]
adata.obs.head()
```

## 4.NORMALIZATION AND LOG TRANSFORMATION

```python
sc.pp.normalize_total(adata, target_sum=1e4)
```

## 5. Log-transform (variance stabilization)

```python
sc.pp.log1p(adata)
```

## 6. Identify highly variable genes (HVGs)

```python
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var.highly_variable]
adata.var.head()
```

## 7. Scale and regress out confounders

```python
sc.pp.scale(adata, max_value=10)
```

## 8. PCA

```python
sc.tl.pca(adata, svd_solver="arpack")
sc.pl.pca_variance_ratio(adata, log=True)
```

## 9. Compute neighborhood graph

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
```

## 10. UMAP embedding for visualization

```python
sc.tl.umap(adata)
sc.pl.umap(adata, color=["n_genes_by_counts", "pct_counts_mt"])
```

## 11. Leiden clustering

```python
sc.tl.leiden(adata, resolution=0.5)
sc.pl.umap(adata, color=["leiden"])
```

## 12. Find marker genes for each cluster

```python
sc.tl.rank_genes_groups(adata, "leiden", method="wilcoxon")
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

## 13. Annotate clusters

```python
sc.get.rank_genes_groups_df(adata, group="0").head(20)
```

```python
import pandas as pd
# make sure you've run rank_genes_groups already:
# sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')

# Collect top 20 markers per cluster into dataframes
from scanpy import get
clusters = adata.obs['leiden'].cat.categories.tolist()
topn = 20

top_markers = {}
for cl in clusters:
    df = sc.get.rank_genes_groups_df(adata, group=cl)
    top_markers[cl] = df['names'].tolist()[:topn]

# show for cluster "0"
print("Top markers for cluster 0:", top_markers.get('0')[:10])
# to inspect all clusters quickly:
for cl, genes in top_markers.items():
    print(f"Cluster {cl} top genes:", genes[:8])
```

```python
# replace var_names with gene symbols
adata.var_names = [adata.var.loc[gene_id, 'gene_symbols'] if 'gene_symbols' in adata.var.columns else gene_id
                   for gene_id in adata.var_names]

# Check the first 10
print(adata.var_names[:10])

```

```python
import pandas as pd

# Create a cross-tab: clusters vs Cell.group
cluster_cellgroup = pd.crosstab(adata.obs['leiden'], adata.obs['Cell.group'])
print(cluster_cellgroup)
```

```python
# For each Leiden cluster, pick the most frequent Cell.group
cluster2celltype = adata.obs.groupby('leiden')['Cell.group'] \
                            .agg(lambda x: x.value_counts().idxmax()) \
                            .to_dict()

# See the mapping
print(cluster2celltype)
```

```python
adata.obs['cell_type'] = adata.obs['leiden'].map(cluster2celltype).astype('category')
sc.pl.umap(adata, color='cell_type', legend_loc='on data')
```
